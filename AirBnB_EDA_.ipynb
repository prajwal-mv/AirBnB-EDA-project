#**What does Airbnb stand for?**
Airbnb is an abbreviation for **A**ir mattress **B**ed and **B**reakfast.


#**What is Airbnb?**
An online marketplace that involves the renting of property to travellers. 

Airbnb does not own any of the properties. It provides a platform on which people can rent out their properties or spare rooms to guests. Prices are set by the property owners and monies are collected via the Airbnb app.

There are many different types of Airbnbs. You can rent a room in someone’s house or a whole island and everything in between!



#**The History of Airbnb**
Airbnb was founded in August 2008 by Joe Gebbia, Brian Chesky, and Nathan Blecharczyk, with its headquarters in San Francisco, USA.

The idea came to Brian Chesky and Joe Gebbia in 2007 when they could not afford to pay the rent for their apartment. They transformed their living room into a bedroom in order to ‘share’ their home with three guests and provided them with homemade breakfast. This was the beginning of the Airbnb concept.

#**Data Preprocessing**

```python
# lets check on the missing values
df.isnull().sum()
```

```python
# Dropping the columns that are not required

df.drop(columns=['name','host_name','latitude','longitude','last_review','calculated_host_listings_count'], inplace = True)
```

If there are Nan values in the reviews per month column that means those Airbnbs have not received the reviews and thus those Nan values can be replaced with 0. 

```python
# Replacing the null values of reviews per month column with 0

df[['reviews_per_month']] = df[['reviews_per_month']].fillna(0)
```

The price of any Airbnb can not be equal to 0, so replacing the 0 priced Airbnbs with the mean of the price column

```python
# Replacing the 0 priced Airbnbs with the median of the price column

df['price'] = df['price'].replace(0,np.median(df['price']))
```

#**Univariate Analysis**


## **Top 10 host IDs that owns the maximum number of Airbnbs in the New York City**


```python
# Creating a dataframe representing the top 10 host IDs who own maximum number of Airbnbs
top_10_host_id = pd.DataFrame({'Host ID':df['host_id'].value_counts(ascending=False).index,
                               'No of Airbnbs':df['host_id'].value_counts(ascending=False).values})[:10]

# Defining the size of the plot
fig,ax=plt.subplots(figsize=(12,7))

# Plotting a bar graph
sns.barplot(x='Host ID', y='No of Airbnbs', data = top_10_host_id).set(title='Host IDs that owns maximum number of Airbnbs')
plt.show()
```

## **Distribution of the Airbnbs in the neighbourhood groups**

```python
#Creating data frame based on the number of stays in each boroughs
Count_grp_neighbourhood = df['neighbourhood_group'].value_counts().to_frame().reset_index()
Count_grp_neighbourhood.columns = ['neighbourhood_group','Count']

#Visualizing the Boroughs
plt.rcParams["figure.figsize"] = [7, 3]
plt.rcParams["figure.autolayout"] = True
Labels1 = list(zip(list(Count_grp_neighbourhood.neighbourhood_group),list(Count_grp_neighbourhood.Count)))
patches, texts = plt.pie(list(Count_grp_neighbourhood.Count))
plt.legend(patches, Labels1, loc="best")
plt.title('Bookings in City ', size=15, color='#4f4e4e')
plt.axis('equal')
```

## **Top 20 neighbourhoods who received the maximum number of reviews**

```python
# Creating a dataframe representing the neighbourhoods who received maximum number of reviews 
most_reviewed_neighbour = pd.DataFrame({'Neighbourhood':df.groupby('neighbourhood')['number_of_reviews'].mean().sort_values(ascending=False).index,
                                        'Mean Reviews':df.groupby('neighbourhood')['number_of_reviews'].mean().sort_values(ascending=False).values})[:20]

# Defining the size of the plot
fig,ax=plt.subplots(figsize=(18,8))

# Defining the label and title of the graph
ax.set_xticklabels(most_reviewed_neighbour['Neighbourhood'],rotation=45)
ax.set_title('Top 20 neighbourhoods that received the maximum number of reviews')

# Plotting a bar graph
figure = sns.barplot(x='Neighbourhood',y='Mean Reviews',data=most_reviewed_neighbour)
plt.show(figure)
```

## **Top 10 reviewed host IDs**

```python
# Creating a dataframe representing the neighbourhoods who received maximum number of reviews 
host_id_review = pd.DataFrame({'Host ID':df.groupby('host_id')['number_of_reviews'].max().sort_values(ascending=False).index[:10],
                               'Mean Review':df.groupby('host_id')['number_of_reviews'].max().sort_values(ascending=False).values[:10]})

# Defining the size of the plot
fig,ax=plt.subplots(figsize=(15,8))

# Defining the label and title of the graph
ax.set_xticklabels(labels = host_id_review['Host ID'], rotation=45)
ax.set_title('Top 10 reviewed host IDs')

# Plotting a bar graph
figure = sns.barplot(x='Host ID', y='Mean Review', data = host_id_review)
plt.show(figure)
```
